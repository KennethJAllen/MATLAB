## Delete Data

```python
house_array = np.array(house_img)
m, n = np.shape(house_array)
known_ratio = 0.65 # ratio of known to unknown entries
num_known = round(known_ratio * m * n) #number of known entries
mask = utils.random_mask(m, n, num_known)
house_missing_data = house_array.copy()
house_missing_data[~mask] = 0
house_missing_img = Image.fromarray(house_missing_data)
house_missing_img
```
